```python
import json
from datetime import datetime
fp=open("rec2.json","r")
data=fp.read()
fp.close()
products=json.loads(data)
x="c"
amt=0
id=[]
name=[]
qu=[]
price=[]
typ=[]
com=[]
while(x=='c'):
    p_name=input("enter the name of the product you want to buy")
    for i in products:
        if(products[i]['pname']==p_name):
            p_quant=int(input("enter the quantity you want to buy"))
            if(products[i]['quantity']<p_quant):
                print("sorry required quantity is not available")
                x=input("type c if you want to continue else s")  
                break
            else:  
                amt+=products[i]['price']*p_quant
                products[i]['quantity']-=p_quant
                id.append(i)
                name.append(products[i]['pname'])
                qu.append(p_quant)
                price.append(products[i]['price']*p_quant)
                typ.append(products[i]['Ptype'])
                com.append(products[i]['company'])
                x=input("type c if you want to continue else s")
                break
    else:
        print("sorry the required product is not available")
        x=input("type c if you want to continue shopping else s if you want to stop")
        
print()
x='c'
y=input("type y if you want to remove any items from the list else n for no")

if y=='y':
    while(x=='c'):
        r=input("enter the product name if you don't want to buy it")
        for i in name:
            if i==r:
                ind=name.index(r)
                amt-=price[ind]
                del name[ind]
                break
        del id[ind]
        del qu[ind]
        del price[ind]
        del typ[ind]
        del com[ind]
        x=input("type c if you still want to remove items or else s if you completed removing items")
    
js=json.dumps(products)
fp=open("rec2.json","w")
fp.write(js)
fp.close() 
print("\n\n")
print("BILL")
print("_______________________")
print()
print("id",end="     ") 
print("name",end="       ")
print("quantity",end="    ")
print("company",end="           ")
print("price")
print()

for i in range(len(id)):
    print(id[i],end="   ") 
    print(name[i],end="       ")
    print(qu[i],end="          ")
    print(com[i],end="              ")
    print(price[i])
print()
print()
print("Total bill=",amt)
    
print("______________________")
print()
print("THANK YOU")
print()
print()
now=datetime.now()
user=str(input("Enter user name:"))
uphoneno = input("Enter the phoneno: ")
fp=open("sales.json","a")
fp.write("User name: ")
fp.write(user)
fp.write("   ")
fp.write("Phone number: ")
fp.write(uphoneno)
fp.write("   ")
fp.write("time:")
fp.write(str(now))
fp.write("\n")
fp.close()
for i in range(len(id)):
    sales = {"productid":i,"typeofproduct":typ[i],"company":com[i],"price":price[i],"quantity":qu[i]}
    j=json.dumps(sales)
    fp=open("sales.json","a")
    fp.write(j)
    fp.write("\n")
    fp.close()
fp=open("sales.json","a")
fp.write("Total bill: ")
fp.write(str(amt))
fp.write("\n")
fp.write("\n")
fp.close()             
```
